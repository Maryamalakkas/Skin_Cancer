```python
lsit =['itch','grew','hurt','changed','bleed','elevation','biopsed']
# convert unkown and unk to false 
df[lsit] = df[lsit].replace({'UNKOWN':False,'UNK':False,'NAN':False})

df[lsit] = df[lsit].replace({'TRUE':True,'FALSE':False})
```

```python
import pandas as pd
from sklearn.feature_extraction import FeatureHasher

# Load your dataset
data = df.copy()

# Selecting the categorical columns to be encoded
categorical_cols = ['background_father', 'background_mother','region']

# Initialize the FeatureHasher with a specified number of features
hasher = FeatureHasher(n_features=10, input_type='string')

for col in categorical_cols:
    # Transform the column
    hashed_features = hasher.transform(data[col].astype(str))
    
    # Convert the sparse matrix to a dense DataFrame
    hashed_features_df = pd.DataFrame(hashed_features.toarray())
    
    # Naming the new columns
    hashed_features_df.columns = [f"{col}_hash_{i}" for i in range(hashed_features_df.shape[1])]
    
    # Drop the original column from the dataframe
    data.drop(col, axis=1, inplace=True)

    # Concatenate the new columns to the original dataframe
    data = pd.concat([data, hashed_features_df], axis=1)

# Display the first few rows of the modified DataFrame
print(data.head())
```

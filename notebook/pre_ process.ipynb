# Cleaning the combined data

```python
df = pd.read_csv('../Data/combined_skin_cancer.csv')
df
```

```python
df.info()
```

```python
df[df.select_dtypes(include=['object']).columns] = df.select_dtypes(include=['object']).astype(str).apply(lambda x: x.str.upper())
df
```

```python
# print objects with true and false 
df.select_dtypes(include=['object']).apply(lambda x: x.str.contains('TRUE|FALSE')).sum()
# if sum is > 0 then create a list of columns
```

```python
lsit =['itch','grew','hurt','changed','bleed','elevation','biopsed']
# convert unkown and unk to false 
df[lsit] = df[lsit].replace({'UNKOWN':False,'UNK':False,'NAN':False})

df[lsit] = df[lsit].replace({'TRUE':True,'FALSE':False})
```

```python
def print_unique_col_values(df):
    for column in df:
        # if df[column].dtype == 'object':
        print(f'{column}: {df[column].unique()}')
```

```python
# print unique values in diagnosis
df['diagnostic'].unique()
# REPLACE nAn with healthy
df['diagnostic'] = df['diagnostic'].replace({'NAN':'HEALTHY'})
```
